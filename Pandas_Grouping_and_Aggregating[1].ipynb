## 1.Write a Pandas program to split the following dataframe into groups based on school code. Also check the type of GroupBy object.

```python
import pandas as pd
import numpy as np
student_data = pd.DataFrame({'school_code':['s001','s002','s003','s001','s002','s004'],
                            'class':['V','V','VI','VI','V','VI'],
                            'name':['Alberto Franco','Gino Mcneil','Ryan Parkers','Eesha Hinton','Gino Mceil','David Parkers'],
                            'date_of_birth':['15/05/2002','17/05/2002','16/02/1999','25/09/1998','11/05/1997','15/09/1997'],
                            'age':[12,12,13,13,14,12],
                            'height':[173,192,186,167,151,159],
                            'weight':[35,32,33,30,31,32],
                            },index=['S1','S2','S3','S4','S5','S6'])
```

```python
print("Original DataFrame:")
student_data
```

```python
print("\nSplit the said data on school_code wise:")
result = student_data.groupby(['school_code'])
for name,group in result:
    print("\nGroup")
    print(name)
    print(group)
print("\nType of the object:")
print(type(result))
```

## 11.Write a Pandas program to split the following dataframe into groups and calculate monthly purchase amount.

```python
orders_data = pd.DataFrame({'ord_no':[70001,70009,70002,70004,70007,70005,70008,70010,70003,70012,70011,70013],
                           'purch_amt':[150.5,270.65,65.26,110.5,948.5,2400.6,5760,1983.43,2480.4,250.45,75.29,3045.6],
                            'ord_date':['05-10-2012','09-10-2012','05-10-2012','08-17-2012','10-09-2012','07-27-2012','10-09-2012','10-10-2012','10-10-2012','06-17-2012','07-08-2012','04-25-2012'],                           'customer_id':[3005,3001,3002,3009,3005,3007,3002,3004,3009,3008,3003,3002],
                           'salesman_id':[5002,5005,5001,5003,5002,5001,5001,5006,5003,5002,5007,5001,]})
```

```python
print("Original DataFrame:")
```

```python
orders_data
```

## 12.Write a Pandas program to split the following dataframe into groups, group by month and year based on order date and find the total purchase amount year wise, month wise.

```python

orders_data = pd.DataFrame({'ord_no':[70001,70009,70002,70004,70007,70005,70008,70010,70003,70012,70011,70013],
                           'purch_amt':[150.5,270.65,65.26,110.5,948.5,2400.6,5760,1983.43,2480.4,250.45,75.29,3045.6],
                            'ord_date':['05-10-2012','09-10-2012','05-10-2012','08-17-2012','10-09-2012','07-27-2012','10-09-2012','10-10-2012','10-10-2012','06-17-2012','07-08-2012','04-25-2012'],                           'customer_id':[3005,3001,3002,3009,3005,3007,3002,3004,3009,3008,3003,3002],
                           'salesman_id':[5002,5005,5001,5003,5002,5001,5001,5006,5003,5002,5007,5001,]})
```

```python
print("Original DataFrame:")
orders_data
```

```python
orders_data['ord_date'] = pd.to_datetime(orders_data['ord_date'])
result = orders_data.groupby([orders_data['ord_date'].dt.year,orders_data['ord_date'].dt.month]).agg({'purch_amt':sum})
```

```python
result
```

## 14.Write a Pandas program to split the following dataframe into groups based on all columns and calculate Groupby value counts on the dataframe.

```python
data = pd.DataFrame({'id':[1,2,1,1,2,1,2],
                    'type':[10,15,11,20,21,12,14], 'book':['Math','English','Physics','Math','English','Physics',
                                                          'English']})
```

```python
print("Original DataFrame:")
data
```

```python
result = data.groupby(['id','type','book']).size().unstack(fill_value=0)
```
